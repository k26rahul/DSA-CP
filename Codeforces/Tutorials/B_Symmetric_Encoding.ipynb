# decode the encoded `input_string`

### ðŸ‘‰ Solution of the problem (contest/1974/problem/B) begins from here

```python
input_string = 'serofedsoc'
```

```python
distinct_sorted_letters = sorted(list(set(list(input_string))))
distinct_sorted_letters
```

```python
encoding_map = {}

n = len(distinct_sorted_letters)
for i in range(n):
  key = distinct_sorted_letters[i]
  value = distinct_sorted_letters[n-1-i]
  print(f'{key} -> {value}')
  encoding_map[key] = value

print(f'{encoding_map=}')
```

```python
decoded = ''

for letter in input_string:
  new_letter = encoding_map[letter]
  print(f'{letter} -> {new_letter}')
  decoded += new_letter

print(f'{decoded=}')
```
